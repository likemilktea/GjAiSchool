## 1. 문제 정의
- 유방암 데이터셋을 사용해서 유방암 분류하기

## 2. 데이터 수집

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

## 3. 데이터 전처리

## 4. 탐색적 데이터 분석

## 5. 모델 선택 및 하이퍼 파라미터 튜닝

```python
# 데이터 분리
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(cancer['data'],cancer['target'],random_state = 66)

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()

from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
# Ensemble - Bagging
# RandomForest
from sklearn.ensemble import RandomForestClassifier
forest50 = RandomForestClassifier(n_estimators=50, random_state=5)
forest70 = RandomForestClassifier(n_estimators=70, random_state=5)
forest90 = RandomForestClassifier(n_estimators=90, random_state=5)
```

```python
# Ensemble - Boosting 계열
# AdaBoost 모델
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=100,random_state=5)
```

```python
# 선형 분류모델 - LinearSVC, Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

## 6. 학습

```python
knn.fit(X_train,y_train)
```

```python
dt.fit(X_train,y_train)
```

```python
forest50.fit(X_train,y_train)
forest70.fit(X_train,y_train)
forest90.fit(X_train,y_train)
```

```python
ada.fit(X_train,y_train)
```

```python
lr.fit(X_train,y_train)
```

## 7. 평가 및 결과

```python
# 성능은 낮다.
# 최적화가(일반화) 잘 되어있다. > train과 test의 차이가 적다
# 결론 보다 안정적이다

print(knn.score(X_train,y_train))
print(knn.score(X_test,y_test))
```

```python
# 분류 평가지표 확인하기
from sklearn.metrics import classification_report
pred = knn.predict(X_test) # test 데이터의 예측값 y_test를 비교

print(classification_report(
    y_test,pred
))
# precision 정밀도 recall 재현율
# support 전체 데이터 중 해당 데이터의 분포량
# macro avg
# weight avg - 데이터의 빈도별로 가중치 부여. 1에 더 많은 데이터가 있으므로 더 많은 비중을 둠
```

```python
# 성능은 높다
# 최적화(일반화)는 안 되어있다. > train과 test의 차이가 크다.

print(dt.score(X_train,y_train))
print(dt.score(X_test,y_test))
```

```python
# 성능이 아주 높다
# 최적화(일반화)가 잘 되어있다. > train과 test의 차이가 크다.
# 시간이 오래 걸린다

# n_estimators=50>100>70순으로 성능이 조금 향상
# n_estimators 100이면 규칙이 너무 train에 맞추어져있다 > 규칙이 너무 복잡하다
# n_estimators=70이면 100보다는 규칙이 적다. 규칙이 적당해서 새로운 걸 잘 맞춘다. 일반화가 잘 되었다.
# n_estimators=50이면 규칙이 너무 적어서 잘 못맞추는 현상. 과소적합

print(forest90.score(X_train,y_train))
print(forest90.score(X_test,y_test))
```

```python
print(forest70.score(X_train,y_train))
print(forest70.score(X_test,y_test))
```

```python
print(forest50.score(X_train,y_train))
print(forest50.score(X_test,y_test))
```

```python
# 특정 중요도
# 전체의 합이 1
# 숫자가 클수록 중요한 특성이다.
forest70.feature_importances_
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 앙상블- 랜덤 포레스트는 거의 모든 컬럼을 사용
# 여러개의 Decisiont Tree를 종합하기 때문에 어쩌다가 한 번씩 사용된 컬럼이 작은 값을 차지
# 많이 사용된 건 큰 값을 차지
# 적은 비율로 사용되는 특성들이 존재 > 모든 데이터가 고루 중요할 때 사용

plt.barh(range(30),forest70.feature_importances_)
plt.yticks(np.arange(30),cancer['feature_names'])
plt.show()
```

```python
print(ada.score(X_train,y_train))
print(ada.score(X_test,y_test))
```

```python
# ada모델 성능이 가장 괜찮게 나왔으니까 진짜 성능 확인하기
from sklearn.model_selection import cross_val_score
cross_val_score(ada,X_train,y_train,cv=5).mean()
```

```python
# ada모델 여러번의 Decision Tree를 사용
# 전부다 연결해서 하나로 씀
# 애매하게 쓰여지는 건 없음 > 중요하지 않은 건 사용되지 않는다.
# 데이터를 판단해보고 중요하지 않은 데이터가 있을 때 사용

plt.barh(range(30),ada.feature_importances_)
plt.yticks(np.arange(30),cancer['feature_names'])
plt.show()
```

```python
# train test 차이가 적으므로 안정적인 모델
# C 1(기본값) > 10 : 규제를 약하게 하겠다 > 규칙을 단순하게 잡아라
# C = 0.1 : 규제를 강하게 하겠다 > 규칙을 복잡하게 잡아라

lr = LogisticRegression(C = 10)
lr.fit(X_train,y_train)

print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC()
svm.fit(X_train,y_train)
print(svm.score(X_train,y_train))
print(svm.score(X_test,y_test))
```
