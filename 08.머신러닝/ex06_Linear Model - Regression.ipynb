### 1. 성적 데이터 생성

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame( [[2,20],[4,40],[8,80],[9,90]],
            columns=['시간','성적'],
            index=['지호','지혜','민아','도운'])
data
```

## 2. LinearRegression

```python
from sklearn.linear_model import LinearRegression
# 모델 객체 생성
linear_model = LinearRegression()
```

```python
# 학습
linear_model.fit(data[['시간']],data['성적'])
```

```python
# 예측
linear_model.predict([[10]])
```

```python
## 3. y = 10*x + 0
linear_model.coef_
```

```python
# 편향 (절편)
linear_model.intercept_
```
