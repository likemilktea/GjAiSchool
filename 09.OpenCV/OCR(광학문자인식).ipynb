```python
# 이미지 처리를 통해 글자를 추출해보자!
img4 = cv2.imread('./img02.png')
img4 = cv2.cvtColor(img4,cv2.COLOR_RGB2BGR)
# 이미지 흑백 변환
img4 = cv2.cvtColor(img4,cv2.COLOR_BGR2GRAY)
# 이진 이미지 변환
_, img4 = cv2.threshold(img4,150,255,cv2.THRESH_BINARY)

plt.xticks([])
plt.yticks([])
plt.imshow(img4, cmap='gray')
plt.show()
```

```python
text4=pytesseract.image_to_string(img4)
print(text4)
```
