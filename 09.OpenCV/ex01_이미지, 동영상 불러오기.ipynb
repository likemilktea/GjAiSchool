```python
# 동영상 : 초당 많은 사진을 연속적으로 보기
# 보통 : 초당 30장의 사진으로 이루어져 있음
# 일반인 : 초당 12~15장이면 동영상으로 인식
# 전문가 : 초당 15~20장이면 동영상으로 인식
```

```python
# 캠으로부터 동영상을 가져오기
import cv2
# 비디오 불러오기
# 0 : 컴퓨터에 연결된 카메라
# 동영상 경로 적어주면 동영상에 연결
cap = cv2.VideoCapture(0)
cnt = 0

# 녹화 설정
fps = 30.0 # 녹화할 프레임 수
width = int(cap.get(3)) # 녹화할 영상의 가로 크기(픽셀)
height = int(cap.get(4)) # 녹화할 영상의 세로 크기(픽셀)

# 코덱 설정
# avi > DIVX
# mp4 > MPEG
fcc = cv2.VideoWriter_fourcc(*'MPEG')

# 동영상 설정
out = cv2.VideoWriter('video.mp4',fcc,fps,(width,height) )
record = False

#초당 30장의 사진 인식
while True: 
    # 연결해놓은 카메라에서 사진 가져오기
    ret, frame = cap.read() # 한번 실행할 때마다 한장씩의 사진 가져옴
    # ret : 불러오기 성공하면 True, 실패하면 False
    # frame : 불러온 사진 이미지 값
    if not ret:
        print("비디오 읽기 실패 또는 비디오 모두 읽음")
        cap.release() # 연결해놓은 카메라 or 동영상 닫기
        cv2.destroyAllWindows() # 윈도우에 열어놓은 창 닫기
        break # 더이상 사진을 불러오지 않아도 되므로 while문 종료
    
       
    # 이미지를 불러왔을 때
    cv2.imshow('video', frame)
    k = cv2.waitKey(33) # 33ms마다 사진 넘긱기 = 1초에 약 30장의 사진을 보내주고 있다.
    
    # 종료 버튼 만들기
    # 1(!) 누르면 종료하기
    if k == 49: # 1은 아스키코드로 49를 가지고 있음
        print('동영상을 종료합니다')
        cap.release() # 카메라 닫기
        cv2.destroyAllWindows() # 윈도우 창 닫기
        break # while문 종료
    
    # 캡쳐(2)
    if k == 50:
        cnt+=1
        cv2.imwrite(f'./data/cap{cnt}.png',frame, params = [cv2.IMWRITE_PNG_COMPRESSION, 0])
        print('동영상을 캡쳐합니다')
        
    # 녹화시작(3)
    if k == 51:
        record=True
        print('동영상 녹화를 시작합니다')
    
    # 3번을 누르고 나서 실행
    if record:
        out.write(frame)
        
    # 녹화종료(4)
    if k ==52:
        out.release()
        print('동영상 녹화를 종료합니다')
```

```python
# opencv > cv2
import cv2
import matplotlib.pyplot as plt
# 이미지 불러오기
img = cv2.imread('./data/funny.jpeg')
# 색상바꾸기
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# 이미지에 표현하기
# 사각형 : cv2.rectangle(배경이미지, 좌상단점좌표, 우하단점좌표, 색상, 선두깨)
# 글씨쓰기 : cv2.putText(배경이미지, 출력내용, 출력시작좌표,폰트,크기,색상,굵기)

import cv2
import matplotlib.pyplot as plt
# 이미지 불러오기
img = cv2.imread('./data/funny.jpeg')
img = cv2.rectangle(img,(30,75), (150,200),(0,0,255),3)
img = cv2.putText(img,'Fire',(150,200),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
cv2.imshow('rec',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
