### Crawling
- 데이터 수집

##### requests 라이브러리
- 서버에게 데이터를 요청하여 응답받기 위한 라이브러리
- 원래는 브라우저를 통해서 서버에 페이지 정보를 요청하지만, 크롤링에서는 requests라이브러리로 페이지정보 요청 후 응답을 받는다!

```python
# 라이브러리 불러오기
import requests as req
# 페이지 정보를 요청하고 응답 받으면 역할 끝
```

```python
# 네이버 페이지의 정보 요청 후 응답 받기
url = 'https://www.naver.com/'# url 주소 : 문자열 형태로 저장
res = req.get(url) # url 주소의 페이지 정보 가져와!
res
```

- Response [200] : 성공적으로 응답받았다
- Response [400] : 클라이언트의 요청에 문제가 있을 때 발생하는 오류 코드
- Response [500] : 서버의 응답에 문제가 있을 대 발생하는 오류 코드

```python
# 우리가 요청한 naver 페이지의 정보 확인
res.text
```

### '날씨'를 검색한 페이지에서 데이터 수집
- 페이지가 변경되면 반드시 변경된 페이지 정보를 다시 요청 응답!!

```python
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%82%A0%EC%94%A8'
res = req.get(url)
res
```

```python
res.text
# 현재 우리가 받아온 정보는 문자열 형태!!
# 컴퓨터가 인지할 수 있도록 만들어줘야 한다!!
```

##### BeautifulSoup 라이브러리
- 컴퓨터가 알아먹을 수 있는 html 언어로 변경해주는 라이브러리

```python
!pip install bs4
```

```python
from bs4 import BeautifulSoup as bs
```

```python
# beautifulsoup 객체화 --> 파싱(컴퓨터가 이해하도록 만드는 것)
# bs(페이지 정보, 파싱 방법)
soup = bs(res.text,'lxml')
soup
```

```python
soup.select('a')
# '통합' 글씨를 가지는 a 태그 선택
# 모든 a태그를 가져옴
```

```python
# soup.select('태그 선택자 이름')
soup.select('a.tab')[0].text
# a.tab을 가지는 데이터가 여러개이기 때문에 리스트에 담겨있음!
# 이 중 첫 번째 데이터만 확인하기 위해 인덱싱
```

```python
# 우리가 추출해온 데이터 -> 요소
# 필요한 데이터는 -> 콘텐츠
# 요소에서 콘텐츠 추출하는 방법 : 요소.text
soup.select('a.tab')[0].text
```

- id 선택자 : (#) 전체 페이지에서 단 한 번만 사용
- class 선택자 : (.) 전체 페이지에서 여러번 사용 가능

```python
# .text는 요소에 적용가능한 속성!! 리스트에 적용 금지!!
soup.select('a.tab').text
```
