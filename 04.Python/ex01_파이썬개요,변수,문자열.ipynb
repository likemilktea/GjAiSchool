```python
# % 
#  :     
#       .->   .
month = 6
day = 13
s1 = " %d %d ."%(month,day) #   
print(s1)
```

```python
# format 
#  :       
#  :   ->    
month = 6
day = 13
s2 = " {} {} .".format(month,day)
print(s2)
```

```python
# f 
#  : .
month = 6
day = 13
s3 = f" {month} {day} ."
print(s3)
```

```python
x=100
y=200
add=x+y
print("%d %d  %d."%(x,y,add))
print("{} {}  {}.".format(x,y,add))
print(f"{x} {y}  {add}.")
```
