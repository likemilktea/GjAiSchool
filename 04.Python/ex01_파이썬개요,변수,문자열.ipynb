### 문자열 인덱싱, 슬라이싱

- 주의사항
    - 띄어쓰기도 인덱스 번호를 가진다!
    - 인덱스 번호는 0번부터 시작한다!<br><br>

- 인덱싱 - 문자열 안에서 내가 원하는 데이터를 하나 뽑아올 때 사용!
    - 변수명[인덱스 번호]<br><br>
    
  
- 슬라이싱 - 문자열 안에서 범위를 주고 여러개의 데이터를 가져올 때 사용!
    - 변수명[인덱스 범위]

```python
# 인덱싱
# 인덱싱 방법: 변수명[인덱스 번호]

name = "My name is sol"

# M 가져오기
print(name[0])
# 인덱스 번호는 음수도 가진다!
print(name[-3:])
```

```python
# 슬라이싱
# 슬라이싱 방법 : 변수명[시작인덱스:종료인덱스 +1]
print(name[3:7])
```

```python
# My 가져오기
print(name[:2])
# 처음부터 가져오고 싶다면 콜론 앞의 인덱스를 비워두면 된다!

# 이름 가져오기
print(name[-3:])
# 끝까지 가져오고 싶다면 콜론 뒤의 인덱스를 비워두면 된다!

# 전체 가져오기
print(name)
```

```python
yd_info="20230627Sunny"
year=yd_info[:4]
day=yd_info[4:8]
weather=yd_info[8:]

#파이썬에서 문자열을 합칠 때에는 '+'기호 사용하기!
print("연도 : "+year)
print("월일 : "+day)
print("날씨 : "+weather)
print(f"{year}{day}{weather}")
print("%s%s%s"%(year,day,weather))
```

### 문자열 포매팅
- 문자열 내부에서 특정한 값(변수)을 삽입해서 출력하고 싶을 때 사용
    - %기호 포매팅
        - 단점 : 자료형에 따라 다른 알파벳 사용
    - format 함수 포매팅
    - f문자열 포매팅 (직관적이다)


```python
# %기호 포매팅
# 단점 : 파이썬에서 자료형에 있어 자유로웠는데 
# 각 자료형에 따라서 다른 알파벳을 써줘야 한다.-> 암기해야하는 단점이 있다.
month = 6
day = 13
s1 = "오늘은 %d월 %d일 입니다."%(month,day) # 앞에서부터 순서대로 들어간다
print(s1)
```

```python
# format 함수
# 장점 : 자료형에 따라서 다른 기호를 사용하지 않아도 된다
# 단점 : 직관적이지 않다 -> 변수가 많아지만 짝짓기가 힘듦
month = 6
day = 13
s2 = "오늘은 {}월 {}일 입니다.".format(month,day)
print(s2)
```

```python
# f문자열 포매팅
# 장점 : 직관적이다.
month = 6
day = 13
s3 = f"오늘은 {month}월 {day}일 입니다."
print(s3)
```

```python
x=100
y=200
add=x+y
print("%d와 %d의 합은 %d입니다."%(x,y,add))
print("{}와 {}의 합은 {}입니다.".format(x,y,add))
print(f"{x}와 {y}의 합은 {add}입니다.")
```

```python
# replace 함수!
s_name = 'My name is sol'
s_name = s_name.replace("sol","☀️")
print(s_name)
```
