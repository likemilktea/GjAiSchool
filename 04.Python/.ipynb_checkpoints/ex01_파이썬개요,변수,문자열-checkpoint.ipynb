### 파이썬 주피터노트북 사용법
- 셀 하나를 통째로 코드영향을 받지 않는 기록형태로 만들고 싶다면 마크다운 모드로 변환
    -command mode에서 'm'을 누르면 된다!(앞의 대괄호가 사라지면 된 것!)
-마크다운 셀이 되면 이미지 넣기도 가능!



- command mode - 앞쪽에 띠가 파란색일 때 (셀 단위의 작업을 할 수 있는 모드!)
    - a : 위쪽에 셀 추가
    - b : 아래쪽에 셀 추가
    - dd : 셀을 삭제
    - m : command mode일 때 해당 셀을 markdown 모드로 만드는 방법
- edit mode - 앞쪽에 띠가 초록색일 때 (셀 내부에 코드를 편집할 수 있는 모드!)
- 실행 단축키
    - ctrl + enter (현재 셀 실행)
    - alt + enter (현재 셀 실행 후 아래 새로운 셀 생성)
    - shift + enter (현재 셀 실행 후 커서를 아래로 이동)

### 문자열 인덱싱, 슬라이싱

- 주의사항
    - 띄어쓰기도 인덱스 번호를 가진다!
    - 인덱스 번호는 0번부터 시작한다!

- 인덱싱 - 문자열 안에서 내가 원하는 데이터를 하나 뽑아올 때 사용!
    - 변수명[인덱스 번호]
    
  
- 슬라이싱 - 문자열 안에서 범위를 주고 여러개의 데이터를 가져올 때 사용!
    - 변수명[인덱스 범위]

```python
# 인덱싱
# 인덱싱 방법: 변수명[인덱스 번호]

name = "My name is sol"

# M 가져오기
print(name[0])
# 인덱스 번호는 음수도 가진다!
print(name[-3:])
```

```python
# 슬라이싱
# 슬라이싱 방법 : 변수명[시작인덱스:종료인덱스 +1]
print(name[3:7])
```
