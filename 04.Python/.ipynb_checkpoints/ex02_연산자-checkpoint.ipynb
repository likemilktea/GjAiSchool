```python
#     : input()
#   !!

num1 = int(input(" >>"))
num2 = int(input(" >>"))
print(f"  : {num1+num2}")
print(f"  : {num1-num2}")
print(f"  : {num1*num2}")
print(f"  : {num1/num2}")
```

```python
pytonScore=int(input("python   >>"))
machineScore=int(input("   >>"))
deepScore=int(input("   >>"))
sumScore=sum((pytonScore,machineScore,deepScore))
avgScore=sumScore/3

print(f" : {sumScore}")
print(f" : {avgScore}")
```

```python
time = int(input("  >> "))
hour, time = time // 3600, time % 3600
minute, time = time // 60, time % 60
second = time
print(f"{hour} {minute} {second}")
```
