#### if else문
- if 조건식:
- (들여쓰기) 실행문장1
- else:
- (들여쓰기) 실행문장2
- 주의사항
    - if와 else는 같은 선상에 둬야한다
    - else는 참이 아니었을 경우를 모두 포함하기 때문에 따로 조건식 작성XX


```python
if False:
    print("실행문장1")
else:
    print("실행문장2")
    
# 위의 단순 if문과의 차이점은 조건식이 거짓일 때 그냥 끝내는가, 
# 다른 실행 문장을 출력하고 끝내는가의 차이이다.
```

```python
money = 9000
if money>=10000:
    print("택시를 탄다")
else:
    print("버스를 탄다")
```

```python
num = int(input("정수 입력 >> "))
if num%3==0 and num%5==0:
    print("3과 5의 배수입니다")
else:
    print("3과 5의 배수가 아닙니다")
```

#### if elif else문
- 여러개의 조건이 존재할 때 사용


```python
# 돈이 10000원 이상이라면 택시를, 5000원 이상이라면 버스를 타고 5000원 미만이라면 걸어가세요.
inputNum=int(input("소지금을 입력하세요 >> "))
result = ""
if inputNum >= 10000:
    result = "택시를 탑니다🚕"
elif inputNum >= 5000:
    result = "버스를 탑니다🚌"
else:
    result = "걸어 갑니다👞"
print(result)
```

```python
num1=int(input("첫번째 정수 입력 >> "))
num2=int(input("두번째 정수 입력 >> "))
result = ""
if num1>num2:
    print("첫번째 정수가 더 큽니다.")
elif num1<num2:
    print("두번째 정수가 더 큽니다.")
else:
    print("두 수가 똑같습니다.")
```

```python
score=int(input("점수 입력 >> "))
grade = ''
if 100>= score >= 0:
    if 100 >= score >= 90:
        grade = "A"
    elif 90 > score >= 80:
        grade = "B"
    elif 80 > score >= 70:
        grade = "C"
    elif 70 > score >= 60:
        grade = "D"
    else:
        grade = "F"
    print(f"{score}은(는) {grade}학점 입니다.")
else: print("100점 이내로 입력해주세요.")
```

```python
money = int(input("insert coin >> "))
menu = input("메뉴 선택 >> ")

if menu == '1' or menu == "콜라":
    menu = 600
elif menu == '2' or menu == "사이다":
    menu = 800
elif menu == '3' or menu == "물":
    menu = 1000
else:
    print("메뉴를 잘못 입력하셨습니다.")
    menu = 0
    

cash = money - menu
cash10000 = 0
cash5000 = 0
cash1000 = 0
cash500 = 0
cash100 = 0

if cash > 0 :
    if cash >= 10000:
        cash10000,cash = cash//10000,cash%10000
    if cash >= 5000:
        cash5000,cash = cash//5000,cash%5000
    if cash >= 1000:
        cash1000,cash = cash//1000,cash%1000
    if cash >= 500:
        cash500,cash = cash//500,cash%500    
    cash100 = cash
    print(f"만원 {cash10000}개, \n오천원 {cash5000}개, \n천원 {cash1000}개, \n오백원 {cash500}개, \n백원 {cash100}개")
else:
    print("금액이 모자랍니다")
```
