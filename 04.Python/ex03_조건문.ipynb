### 조건문
- if
- elif
- else

 #### if
 - if 조건식(논리형결과) :
 - (들여쓰기!)실행문장


```python
if True:
    print("실행")
```

```python
# if문 조건식의 결과가 False일 경우 아무것도 실행하지 않고 끝낸다.

if False:
    print("실행")
```

```python
# 들여쓰기의 중요성

if False:
    print("실행")
print("실행2")
```

```python
money = 9000
if money>=10000:
    print("택시를 탄다")
else:
    print("버스를 탄다")
```

#### if else문
- if 조건식:
- (들여쓰기) 실행문장1
- else:
- (들여쓰기) 실행문장2
- 주의사항
    - if와 else는 같은 선상에 둬야한다
    - else는 참이 아니었을 경우를 모두 포함하기 때문에 따로 조건식 작성XX


```python
if False:
    print("실행문장1")
else:
    print("실행문장2")
    
# 위의 단순 if문과의 차이점은 조건식이 거짓일 때 그냥 끝내는가, 
# 다른 실행 문장을 출력하고 끝내는가의 차이이다.
```

```python
money = 9000
if money>=10000:
    print("택시를 탄다")
else:
    print("버스를 탄다")
```

```python
num = int(input("정수 입력 >> "))
if num%3==0 and num%5==0:
    print("3과 5의 배수입니다")
else:
    print("3과 5의 배수가 아닙니다")
```
