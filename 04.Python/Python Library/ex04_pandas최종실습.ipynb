## 범죄 데이터 현황 분석

```python
import pandas as pd
```

```python
# 1.데이터 3개를 불러오기
```

```python
data2019 = pd.read_csv("2019.csv", encoding = 'euc_kr', index_col = "관서명")
data2020 = pd.read_csv("2020.csv", encoding = 'euc_kr', index_col = "관서명")
data2021 = pd.read_csv("2021.csv", encoding = 'euc_kr', index_col = "관서명")
```

```python
# 2019,2020,2021 각각 데이터들의 크기 확인

print(data2019.shape)
print(data2020.shape)
print(data2021.shape)
```

```python
# 인덱스 중 중복을 제거한 한 후 비교하기
data2019.index.unique()
```

```python
data2021.index.unique()
```

```python
# 2.특정 연도에만 존재하고 있는 경찰청 삭제
# 특정연도(2021)에만 존재하는 관서명인 '광주지방경찰청' 행 삭제
data2021.drop('광주지방경찰청', inplace = True)
```

```python
data2021.index.unique()
```

```python
# 3.연도별 데이터 총합 구하기
data2019['총합'] = data2019.loc[:,"살인":"폭력"].sum(axis=1)
data2020['총합'] = data2020.loc[:,"살인":"폭력"].sum(axis=1)
data2021['총합'] = data2021.loc[:,"살인":"폭력"].sum(axis=1)
```

```python
data2019
```

```python
# 4-1 구분이 '발생건수'인 값들만 가져오기 (구분 == 발생건수)
temp_data2019 = data2019[data2019.loc[:,'구분']=='발생건수']
temp_data2020 = data2020[data2020.loc[:,'구분']=='발생건수']
temp_data2021 = data2021[data2021.loc[:,'구분']=='발생건수']
```

```python
temp_data2021
```

```python
# 4-2 총합을 기준으로 데이터 자르기 (Series 형태로 자르기)
s2019 = temp_data2019['총합']
s2020 = temp_data2020['총합']
s2021 = temp_data2021['총합']
```

```python
s2019.name = '2019총계'
s2020.name = '2020총계'
s2021.name = '2021총계'
s2019
```

```python
# 5.증감률 계산
# (금년-작년)/작년*100
plus2020 = (s2020-s2019)/2019*100
plus2021 = (s2021-s2020)/2010*100
```

```python
plus2020.name = '19~20년 증감률'
plus2021.name = '20~21년 증감률'
```

```python
print(plus2020)
```

```python
# 6. 데이터프레임 병합(concate() 함수 사용)
result = pd.concat([s2019,plus2020,s2020,plus2021,s2021],axis=1)
```
