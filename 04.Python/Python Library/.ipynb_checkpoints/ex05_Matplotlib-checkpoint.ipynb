```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# line chart 그리기
y=[2,4,6,8]
plt.plot(y)
# 하나의 데이터만 넣어주면 y축으로 간다. 
# x축은 인덱스 번호로 자동 지정이 된다.
```

```python
x= [1,2,3,4]
y= [2,4,6,8]
# 데이터 형식->list 앞이 x축 뒤가 y축
plt.plot(x,y,ls='-',color='m', marker = 'o',mfc='blue')

plt.show()

# 속성값
# 라인의 스타일 설정 - ls(LineStyle)
# 라인의 두께 지정 = lw(LineWidth)
# 라인의 컬러 지정 - color
# 값의 위치 설정 - marker
# marker의 색 지정 - mfc(MarkerFaceColor)
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]

plt.plot(x,y,ls='-.',lw=5, color='y', marker = '*', ms= 20, mfc='white', mew=4 mec='skyblue')

plt.show()
```
